```python
import pandas as pd
Amazon = pd.read_csv('Amazon Customer Behavior Survey.CSV', header=0)
Amazon
```

```python
selected_columns = [
    'Purchase_Frequency',
    'Product_Search_Method',
    'Customer_Reviews_Importance',
    'Search_Result_Exploration',
    'Rating_Accuracy ',
    'Shopping_Satisfaction'
    
]
table = Amazon[selected_columns]
table
 
```

```python
table['Satisfaction_Label'] = table['Shopping_Satisfaction'].apply(lambda x: 1 if x >= 3 else 0)
table
```

```python
for col in table.columns:
    print(table[col].value_counts())
```

```python
frequency_mapping = {
    'Multiple times a week': 5,
    'Once a week': 4,
    'Few times a month': 3,
    'Once a month': 2,
    'Less than once a month': 1
}
table['Purchase_Frequency'] = table['Purchase_Frequency'].map(frequency_mapping)
table
```

```python
category_t = table.select_dtypes('object')
final_table = pd.get_dummies(category_t)
final_table
```

```python
final_table['Purchase_Frequency']=table['Purchase_Frequency']
final_table['Customer_Reviews_Importance']=table['Customer_Reviews_Importance']
final_table['Rating_Accuracy ']=table['Rating_Accuracy ']
final_table['Shopping_Satisfaction']=table['Shopping_Satisfaction']
final_table['Satisfaction_Label']=table['Satisfaction_Label']
final_table
```

```python
final_table.columns
```

```python
X = final_table.drop(columns=['Shopping_Satisfaction', 'Satisfaction_Label'])
y = final_table['Satisfaction_Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('logreg.coef:', logreg.coef_)
print(logreg.intercept_)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:{}".format(accuracy_score(y_test, y_pred)))
print("Precision:{}".format(precision_score(y_test, y_pred)))
print("Recall:{}".format(recall_score(y_test, y_pred)))
print("F1:{}".format(f1_score(y_test, y_pred)))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
print("SVM Model Metrics:")
print("Accuracy: {}".format(accuracy_score(y_test, y_pred_svm)))
print("Precision: {}".format(precision_score(y_test, y_pred_svm)))
print("Recall: {}".format(recall_score(y_test, y_pred_svm)))
print("F1: {}".format(f1_score(y_test, y_pred_svm)))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=1)
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
print("RBF Kernel SVM Metrics:")
print("Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_rbf)))
print("Precision: {:.2f}".format(precision_score(y_test, y_pred_rbf)))
print("Recall: {:.2f}".format(recall_score(y_test, y_pred_rbf)))
print("F1 Score: {:.2f}".format(f1_score(y_test, y_pred_rbf)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
tree_model = DecisionTreeClassifier(max_depth=5, random_state=1)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
print("Decision Tree Metrics:")
print("Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_tree)))
print("Precision: {:.2f}".format(precision_score(y_test, y_pred_tree)))
print("Recall: {:.2f}".format(recall_score(y_test, y_pred_tree)))
print("F1 Score: {:.2f}".format(f1_score(y_test, y_pred_tree)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
print("Random Forest Metrics:")
print("Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_rf)))
print("Precision: {:.2f}".format(precision_score(y_test, y_pred_rf)))
print("Recall: {:.2f}".format(recall_score(y_test, y_pred_rf)))
print("F1 Score: {:.2f}".format(f1_score(y_test, y_pred_rf)))
```
